```python
import pandas as pd

# ./auto-mpg.csv 파일을 불러와서 데이터프레임으로 생성(헤더는 제외)
df = pd.read_csv('./04auto-mpg.csv', header=None)

# 열 이름을 지정 - ['mpg','cylinders','displacement','horsepower','weight', 'acceleration','model year','origin','name']
df.columns = ['mpg','cylinders','displacement','horsepower','weight', 'acceleration','model year','origin','name']
df
```

```python
# 데이터프레임 df의 내용 확인 

df.info()
```

```python
# 데이터프레임 df의 자료형 확인 

df.dtypes                             # 자료형들 확인이여서 s 붙음. (dtypes)
```

```python
# 시리즈(mpg 열)의 자료형 확인(컬럼명 참조 활용)

# df.mpg.dtype                          # 하나의 자료형 확인이라 dypte
df['mpg'].dtype                         #  == df.mpg.dtype
```

```python
# 데이터프레임 df의 기술통계 정보 확인 - 수치 자료(int, float)에 대해서만 보여줌

df.describe()                           # 수치 자료에 대해 (갯수, 평균, 표준편차, 최소값, 최대값, 4분위수) 보여줌
```

```python
# 행렬 치환하여 파악하기(전치행렬) - 두 가지 방법

# df.describe().transpose()                         # 행과 열을 뒤바꿈(치환)
df.describe().T                                     # df.describe().transpose() == df.describe().T 
```

```python
# 모든 자료에 대한 기술통계 정보 확인 - Object형(고유값, 최대값 등 수치없이 통계가 가능한 정보 확인 가능)

df.describe(include='all')                       # 수치자료형 뿐만 아니라 모든 자료를 다 확인할 수 있음.(include='all')
```

```python
# 데이터프레임 df의 각 열이 가지고 있는 원소 개수 확인 

df.count()
```

```python
# df.count()가 반환하는 객체 타입 출력

type(df.count())
```

```python
# 데이터프레임 df의 특정 열(origin)이 가지고 있는 고유값 갯수 확인(value_counts())

unique_values = df['origin'].value_counts()        # 해당 열의 value_counts 확인 
unique_values                                      # df['origin'].value_counts() == df.origin.value_counts()
```

```python
# value_counts 메소드가 반환하는 객체 타입 출력

type(unique_values)
```

```python
# 데이터프레임의 각 컬럼별 평균값 (numeric_only=True 숫자에 대해서만 평균을 구함)

df.mean(numeric_only=True)                                 # mean : 각 열에 대한 평균, numeric_only=True 숫자들만 평균 구함
```

```python
# 데이터프레임에서 특정 열(mpg)의 평균 - 열 인덱스 사용

df['mpg'].mean()                                            # 하나의 열 선택 후 평균을 구함
```

```python
# 데이터프레임에서 특정 열(mpg)의 평균 - 열 참조 사용

df.mpg.mean()
```

```python
# 두 개 이상의 컬럼('mpg','weight')의 평균 확인(선택할 컬럼을 리스트로 전달)

df[['mpg','weight']].mean()                                 # 리스트로 묶어서 값을 df에 전달
```

```python
# 데이터프레임의 각 컬럼별 중간값(median)

df.median(numeric_only=True)                                # 중간값 : median
```

```python
# 특정 컬럼(mpg)의 중간값

df['mpg'].median()
```

```python
# 데이터프레임의 각 컬럼별 최대값(max())

df.max()                                                    # 최대값 : max()
```

```python
# 특정 컬럼(mpg)의 최대값

df['mpg'].max()
```

```python
# 데이터프레임의 각 컬럼별 최소값 

df.min()
```

```python
# 데이터프레임의 특정 열(mpg)의 최소값

df['mpg'].min()
```

```python
# 데이터프레임의 각 컬럼별 표준편차(std())  - 숫자만 구하도록 파라미터 추가

df.std(numeric_only=True)                                  # 표준편차 : std()
```

```python
# 데이터프레임의 특정 열(mpg)의 표준편차

df['mpg'].std()                                    
```

```python
# 데이터프레임의 각 컬럼간 상관 계수(correlation coefficient)
 
df.corr(numeric_only=True)                                # 상관 계수(서로의 상관 정도) -> .corr()
```

```python
# 특정 컬럼('mpg','weight') 간 상관계수(correlation coefficient) - 선택할 컬럼을 리스트로 전달

df[['mpg','weight']].corr()               # 리스트로 묶어서 전달
```
