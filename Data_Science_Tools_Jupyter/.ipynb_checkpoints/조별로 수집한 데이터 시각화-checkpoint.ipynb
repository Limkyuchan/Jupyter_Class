```python
import pandas as pd

book = pd.read_csv('./KyoboBestSeller.csv', encoding='CP949')
book.head()
```

```python
a = book.iloc[0:21]
a
```

```python
book['순 판매가격'] = book['판매가격'].str[:-1]
book
```

```python
book.fillna(0, inplace=True)
book
```

```python
book['가격2'] = book['순 판매가격'].str[:-4] + book['순 판매가격'].str[-3:] 
book['가격2']   
```

```python
book['가격2'] = book['가격2'].astype('int')
```

```python
book['순 판매가격'].replace(","," ")
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = "./06malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

axe1 = plt.subplot()
axe2 = axe1.twinx()

c1 = axe1.plot(a.index, book['평점'].iloc[1:21], color='r')
c2 = axe2.plot(a.index, book['가격2'].iloc[1:21], color='b')

c = c1 + c2
axe1.legend(c, ['평점','판매가격'])

axe1.set_ylabel('평점')
axe2.set_ylabel('판매가격')
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_path = "./06malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

ax1 = plt.subplot()

ax1.plot(a.index, book['평점'].iloc[1:21], '-s', color='hotpink', label='평점')
ax1.set_xticks(a.index)   # x축 1 ~ 20위까지 1 단위로 표시
ax1.set_ylim(9,10)
ax1.set_xlabel('베스트셀러 20위')
ax1.set_ylabel('평점')

ax2 = ax1.twinx()
ax2.bar(a.index, book['가격2'].iloc[1:21],color='lightblue', label='판매가격')
ax2.set_xticks(a.index)   # x축 1 ~ 20위까지 1 단위로 표시
ax2.set_ylim(0,40000)
ax2.set_ylabel('판매가격')

ax1.set_zorder(ax2.get_zorder() + 10)
ax1.patch.set_visible(False)

ax1.legend(loc='upper left')   
ax2.legend(loc='upper right')

plt.show()
```
