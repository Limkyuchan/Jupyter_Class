```python
import numpy as np
import pandas as pd

iris = pd.read_csv("IRIS.csv")
iris.info()

iris_data = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].to_numpy()
iris_data
```

```python
iris_target = iris['species'].to_numpy()
iris_target
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters= 3, random_state=42, n_init='auto')

km.fit(iris_data)
```

```python
len(km.labels_), km.labels_
```

```python
iris[km.labels_ == 1]
```

```python
iris[km.labels_ == 2]
```

```python
iris[km.labels_ == 0]
```

```python
import matplotlib.pyplot as plt
inertia = []

for k in range(2, 7) :
    km = KMeans(n_clusters=k, random_state=42, n_init='auto')
    km.fit(iris_data)
    inertia.append(km.inertia_)
    
plt.plot(range(2,7), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```
